```python
import torch
import os
import glob
import torch
from tqdm import tqdm
import scipy.io.wavfile as wav
import numpy as np
```

```python
params = {
    'root_dir': 'unifieddataset',
    'feat_dir': 'unifieddataset/features',
    'sampling_rate': 24000,
    'hop_length_s': 0.02,
    'nb_mels': 64
}
```

```python
split = 'dev'

def load_audio(audio_file, fs):
        fs, audio = wav.read(audio_file)
        audio = audio[:, :4] / 32768.0 + 1e-8 # up to 4 channels
        return audio, fs
```

```python
audio_files = glob.glob(os.path.join(params['root_dir'], 'audio_dev', 'dev-*', '*.wav'))
os.makedirs(os.path.join(params['feat_dir'], f'audio_{split}'), exist_ok=True)

for audio_file in tqdm(audio_files, desc=f"Processing audio files ({split})", unit="file"):
    filename = os.path.splitext(os.path.basename(audio_file))[0] + '.pt'
    feature_path = os.path.join(params['feat_dir'], f'audio_{split}', filename)
    # Check if the feature file already exists
    if os.path.exists(feature_path):
        continue
    # If the feature file doesn't exist, perform extraction
    audio, sr = load_audio(audio_file, params['sampling_rate'])
    # audio_feat = extract_log_mel_spectrogram(audio, sr, self.n_fft, self.hop_length, self.win_length, self.nb_mels)
    # audio_feat = torch.tensor(audio_feat, dtype=torch.float32)
    torch.save(audio, feature_path)
```
